**[3. Trend Identification]**

Movies vs TV Shows

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

Content Added Per Year

```python
df['release_year'].value_counts().sort_index().plot()
```

Most Common Ratings

```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=df,
              order=df['rating'].value_counts().index)

plt.title("Content Ratings Distribution")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

Top 10 Countries Producing Content

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index)

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

Release Year Distribution

```python
plt.figure(figsize=(12,6))
sns.histplot(df['release_year'], bins=30, kde=True)

plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

Top Genres

```python
# Split genres
genres = df['listed_in'].str.split(', ', expand=True).stack()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_genres.values, y=top_genres.index)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```


## Insights from Netflix Dataset EDA

1. Netflix has more Movies than TV Shows in its content library.

2. The number of titles added on Netflix increased rapidly after 2015.

3. The highest growth in content additions was observed between 2017 and 2020.

4. The United States contributes the largest amount of Netflix content.

5. India and the United Kingdom are also major contributors to Netflix titles.

6. TV-MA and TV-14 are the most common content ratings on Netflix.

7. Most Netflix content is targeted toward teenagers and adult audiences.

8. Drama, International Movies, and Comedy are among the most popular genres.

9. Most content available on Netflix was released after the year 2000.

10. Netflix focuses mainly on modern and recently released entertainment content.

11. Correlation analysis showed weak relationships between numerical variables.

12. The dataset contains mostly categorical data rather than numerical data.

13. Overall, Netflix has expanded globally with a strong focus on international and mature audience content.
